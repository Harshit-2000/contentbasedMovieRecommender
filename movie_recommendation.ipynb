```python
import pandas as pd
import numpy as np

import ast

import nltk
from nltk.stem import PorterStemmer
```

```python
movies = pd.read_csv('data/tmdb_5000_movies.csv')
credits = pd.read_csv('data/tmdb_5000_credits.csv')
```

```python
movies_rating = movies[['title', 'id', 'vote_average', 'vote_count']]
```

```python
 movies.head()
```

```python
credits.head()
```

```python
movies = movies.merge(credits, on = 'title')
```

```python
movies.info()
```

```python
movies = movies[['genres', 'id', 'keywords', 'overview', 'title', 'cast', 'crew']]
```

```python
movies.head()
```

```python
def extractName(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
movies['keywords'] = movies['keywords'].apply(extractName)
movies['genres'] = movies['genres'].apply(extractName)
```

```python
def get_cast(obj):
    L = []
    n = 0
    for i in ast.literal_eval(obj):
        if n == 4:
            break
        L.append(i['name'])
        n += 1
    return L
```

```python
movies['cast'] = movies['cast'].apply(get_cast)
```

```python
def get_crew(obj):
    L = []
    for i in ast.literal_eval(obj):
        if (i['job'] == 'Director'):
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(get_crew)
```

```python
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
def remove_stop_words(string):
    words = [word for word in str(string).split() if word.lower() not in ENGLISH_STOP_WORDS]
    return words
```

```python
movies['overview'] = movies['overview'].apply(remove_stop_words)
```

```python
movies['tags'] = movies['genres'] + movies['keywords'] + movies['overview'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
df = movies[['id', 'title', 'tags']]
```

```python
# Apply lower case and remove spaces
def lowerCase_spaces(obj):
    L = []
    for word in obj:
        L.append(word.lower().replace(' ', ''))
    return L
```

```python
df['tags'] = df['tags'].apply(lowerCase_spaces)
```

```python
df.head()
```

```python
df['tags'] = df['tags'].apply(lambda x : ' '.join(x))
```

```python
ps = PorterStemmer()
def stem(text):
    L = []
    for i in text.split():
        L.append(ps.stem(i))
    return ' '.join(L)
      
```

```python
df['tags'] = df['tags'].apply(stem)
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 4500, stop_words = 'english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    L = []
    i = df[df['title'] == movie].index[0]
    distances = similarity[i]
    
    similar = sorted(list(enumerate(distances)), reverse = True, key=lambda x: x[1])
    
    similar_movies = similar[1 : 10]
    for i in similar_movies:
#         print(df.iloc[i[0]].id)
        L.append(df.iloc[i[0]].title)
        
    return L
```

```python
recommend('Avatar')
```
